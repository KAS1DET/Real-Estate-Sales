## **รายละเอียดข้อมูล**

* Serial_Number: หมายเลขซีเรียลของธุรกรรม

* List_Year: ปีที่บันทึกข้อมูล

* Date_Recorded: วันที่บันทึกการขาย

* Town: เมืองทีอสังหาริมทรัพย์ตั้งอยู่

* Address: ที่อยู่ของอสังหาริมทรัพย์

* Assessed_Value: ราคาประเมินโดยผู้ประเมินภาษี

* Sale_Amount: ราคาขายจริง (หน่วยเป็น $)

* Sales_Ratio: อัตราส่วนระหว่างราคาประเมินกับราคาขาย (Assessed_Value / Sale_Amount)

* Property_Type: ประเภทของอสังหาริมทรัพย์ (เช่น Residential, Commercial และอื่นๆ)

* Residential_Type: ประเภทของที่อยู่อาศัย (เช่น Single Family, Condo และอื่นๆ)

* Non_Use_Code: โค้ดพิเศษที่อาจบ่งบอกว่าทรัพย์สินมีเงื่อนไขพิเศษ

* Assessor_Remarks และ OPM_Remarks: ข้อสังเกตเพิ่มเติมจากผู้ประเมิน

* Location: พิกัดภูมิศาสตร์ของอสังหาริมทรัพย์

# เริ่มทำการ cleansing data

```python
# เรียกดูรายละเอียดข้อมูล

df.head(10)
```

```python
df.columns = df.columns.str.replace(' ', '_')
df.columns
```

```python
df.rename(columns={'OPM_remarks':'OPM_Remarks'}, inplace=True)
```

```python
df.head(1)
```

```python
# ทำการ เปลี่ยนข้อมูล ที่ว่างใน Columns 'Non_Use_Code', 'Assessor_Remarks', 'OPM_Remarks' ให้เป็น N/A
# และทำการ เช็คค่าว่างใน Columns 'Non_Use_Code', 'Assessor_Remarks', 'OPM_Remarks' ว่ามีกี่ตัว

df[['Non_Use_Code', 'Assessor_Remarks', 'OPM_Remarks']] = df[['Non_Use_Code', 'Assessor_Remarks', 'OPM_Remarks']].fillna('N/A')
df[['Non_Use_Code', 'Assessor_Remarks', 'OPM_Remarks']].isna().sum()
```

```python
# ทำการนับจำนวนที่มี N/A เพื่อเช็คว่าครบกับจำนวนที่ค่าว่างหรือไม่ โดยค่าเก่า ของ แต่ละ Columns มีค่าดังนี้
# Non_Use_Code: 784178
# Assessor_Remarks: 926401
# OPM_Remarks: 1084598

print(df[df['Non_Use_Code'] == 'N/A'].shape[0])
print(df[df['Assessor_Remarks'] == 'N/A'].shape[0])
print(df[df['OPM_Remarks'] == 'N/A'].shape[0])
```

```python
# ทำการเช็คว่า Sale_Amount ที่เป็น 0 และ Sales_Ratio ที่ไม่เท่ากับ 0 มีหรือไม่ หากมีจะได้นำ Sales_Ratio มาหา Sale_Amount
# ทำการเช็คว่า Assessed_Value ที่เป็น 0 และ Sales_Ratio ที่ไม่เท่ากับ 0 มีหรือไม่ หากมีจะได้นำ Sales_Ratio มาหา Assessed_Value

print(df[(df['Sale_Amount'] == 0) & (df['Sales_Ratio'] != 0)].shape[0])
print(df[(df['Assessed_Value'] == 0) & (df['Sales_Ratio'] != 0)].shape[0])
```

```python
# ทำการหา ว่า Sale_Amount ไม่เท่ากับ 0 เพราะต้องนำ Sale_Amount มาหาร
# ทำการหา Sales_Ratio อีกรอบเพื่อความถูกต้องของข้อมูล โดยนำ Assessed_Value / Sale_Amount

mask = (df['Sale_Amount'] != 0)
df.loc[mask, 'Sales_Ratio'] = (df.loc[mask, 'Assessed_Value'] / df.loc[mask, 'Sale_Amount']).round(2)

df.head()
```

```python
len(df['Non_Use_Code'].unique())
```

```python
# นำ 'Non_Use_Code' ที่ไม่ใช่การซื้อขายจริงๆ ออก

exclude_codes = [
    '02 - Love and Affection', '29 - No Consideration', '22 - Money and Personal Property',
    '21 - Personal Property Exchange', '23 - Zoning', '15 - Government Agency',
    '16 - Charitable Group', '20 - Cemetery', '14 - Foreclosure', '18 - In Lieu Of Foreclosure',
    '30 - Auction', '06 - Portion of Property', '08 - Part Interest', '04 - Correcting Deed',
    '07 - Change in Property', '05 - Deed Date'
]

df_filtered = df[~df['Non_Use_Code'].isin(exclude_codes)]

df = df_filtered.copy()
```

```python
# ลบ Sale_Amount ที่น้อยกว่า 1000 เพราะว่า ยอดขายที่เกิด น้อยกว่า 1000 ไม่น่าจะเป็นไปได้

df.drop(df[df['Sale_Amount'] < 1000].index, axis=0, inplace=True)
```

```python
# หาค่าที่ปกติใน 'Sale_Ratio'

Q1 = df['Sales_Ratio'].quantile(0.25)
Q3 = df['Sales_Ratio'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

df_filtered = df[df['Sales_Ratio'] <= upper_bound]
df = df_filtered.copy()
```

# เทคนิคการหาข้อมูลทางสถิติ

```python
data = df.copy()
```

```python
data.dtypes
```

```python
cols_to_normalize = ["Assessed_Value", "Sale_Amount", "Sales_Ratio"]

for col in cols_to_normalize:
    x_Min = data[col].min()
    x_Max = data[col].max()
    data[col] = (data[col] - x_Min) / (x_Max - x_Min)
    
data.head()
```

```python
data['Sale_zScore'] = (data['Sale_Amount'] - data['Sale_Amount'].mean()) / data['Sale_Amount'].std()
data.head()
```

```python
cond = data['Sale_zScore'].abs() <= 3
data_filtereded = data[cond]
data_filtereded.sample(5)
```

### วิเคราะห์จำนวนธุรกรรมตามปี

```python
transactions_per_year = data["Date_Recorded"].dt.year.value_counts().sort_index().drop(1999, errors='ignore')

print(transactions_per_year)
```

```python
plt.figure(figsize=(8, 5))

transactions_per_year.plot(kind='bar', color='#2c325c', width=0.9)

plt.xlabel("Year of data recording", fontsize=12)
plt.ylabel("Number of transactions", fontsize=12)
plt.title("The graph shows the number of real estate transactions by year.", fontsize=14)
plt.xticks(rotation=45)

plt.tight_layout()

plt.grid(axis='y', linestyle='--', alpha=0.7)
```

### วิเคราะห์ 10 อันดับเมืองที่มีราคาขายเฉลี่ยสูงที่สุด

```python
average_sale_price = data.groupby('Town')['Sale_Amount'].mean().sort_values(ascending=False)

average_sale_price.head(10)
```

```python
plt.figure(figsize=(8, 5))

average_sale_price.head(10).plot(kind='bar', color='#2c325c')

plt.title('Top 10 Towns with Highest Average Sale Price', fontsize=14)
plt.xlabel('Town', fontsize=12)
plt.ylabel('Average Sale Price ($)', fontsize=12)
plt.grid(axis='y', linestyle='--')
plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

### วิเคราะห์ราคาเฉลี่ยของอสังหาริมทรัพย์จากเมืองที่มีอสังหาริมทรัพย์มากที่สุดเพื่อดูแนวโน้มตามปี

```python
top_town = data['Town'].value_counts().index[0]
top_town_data = data[data['Town'] == top_town]
yearly_avg = top_town_data.groupby('List_Year')['Sale_Amount'].mean()

plt.figure(figsize=(8, 5))
plt.plot(yearly_avg.index, yearly_avg.values, marker='o', linestyle='-', color='#2c325c', label=f'Average Sale Amount in {top_town}')

plt.title(f'Average Sale Amount in {top_town} by Year', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Sale Amount ($)', fontsize=12)
plt.grid(True)
plt.xticks(yearly_avg.index, rotation=45)
plt.legend()
plt.show()
```

### วิเคราะห์แนวโน้มราคาขายในแต่ละปี

```python
yearly_sales = data.groupby('List_Year')['Sale_Amount'].mean()
# yearly_sales

plt.figure(figsize=(8,6))
yearly_sales.plot(kind='line', marker='o', color='#2c325c', label='Average Sale Amount')

plt.title('Trend of Propery Sale Prices Over Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Sale Amount ($)', fontsize=12)
plt.legend()
plt.grid(True)

plt.show()
```
